```python
import requests
from bs4 import BeautifulSoup
import re

webtoontitle = []
webtoonsummary = []
webtoonrating = []

for n in range(1, 11):
    req = requests.get('https://comic.naver.com/genre/bestChallenge.nhn?&page='+ str(n))
    req.encoding = 'utf-8' 
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    title = soup.select('tr > td > div.challengeInfo > h6 > a')
    summary = soup.select('.summary')
    rating = soup.select('div.challengeInfo > div.rating_type > strong')
    
    for dom in title:
        webtoontitle.append(dom.text.strip())
    for dom in summary:
        webtoonsummary.append(dom.text)
    for dom in rating:
        webtoonrating.append(dom.text)
             
print(webtoontitle)
print(webtoonsummary)
print(webtoonrating)
```
