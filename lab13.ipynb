```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
#1
dict_data = {'1분기' : [500, 690, 1100, 1500, 1990, 1020],
             '2분기' : [450, 700, 1030, 1650, 2020, 1600],
             '3분기' : [520, 820, 1200, 1700, 2300, 2200],
             '4분기' : [610, 900, 1380, 1850, 2420, 2550]}
df = pd.DataFrame(dict_data, index = ['2015', '2016', '2017', '2018', '2019', '2020'])
display(df)
```

```python
#2
plt.figure(figsize = (10, 6))
plt.plot(df)
plt.title('시각화 과제1', color = 'purple')
df.index = df.index.map(lambda x : str(x) + '년')
plt.legend(['1분기', '2분기', '3분기', '4분기'])
plt.savefig("output/hw1.png")
plt.show()
```

```python
#3
df.T.plot(figsize = (10,6))
plt.title('시각화 과제2', color = 'red')
plt.savefig("output/hw2.png")
plt.show()
```

```python
#4
ratio = [20.8, 22.8, 26.7, 29.7]
labels = ['1분기', '2분기', '3분기', '4분기']
explode = [0.05, 0.05, 0.05, 0.05]
colors = ['red', 'green', 'darkred','purple']
plt.pie(ratio, labels = labels, explode = explode, autopct = '%.1f%%',
        colors = colors)
plt.savefig("output/hw3.png")
plt.show()
```

```python
#5
ratio = [9.5, 6.4, 22.5, 26.7, 20.5, 14.4]
labels = ['2016년', '2015년', '2020년', '2019년', '2018년', '2017년']
colors = ['green', 'red', 'lightpink', 'yellow', 'purple', 'darkred']
plt.pie(ratio, labels = labels, colors = colors, autopct = '%.1f%%',
        counterclock=False, startangle = 60)
plt.savefig("output/hw4.png")
plt.show()
```
